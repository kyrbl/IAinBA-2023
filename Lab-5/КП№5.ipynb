```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
X_2_poly = PolynomialFeatures(degree=2, include_bias=False).fit_transform(X)
X_3_poly = PolynomialFeatures(degree=3, include_bias=False).fit_transform(X)
X_4_poly = PolynomialFeatures(degree=4, include_bias=False).fit_transform(X)
```

```python
pol2 = LinearRegression().fit(X_2_poly, Y)
pol3 = LinearRegression().fit(X_3_poly, Y)
pol4 = LinearRegression().fit(X_4_poly, Y)
```

```python
r_sq_2 = pol2.score(X_2_poly, Y)
r_sq_3 = pol3.score(X_3_poly, Y)
r_sq_4 = pol4.score(X_4_poly, Y)
print(f"R^2 for all models\nSecond order {r_sq_2}\nThird order {r_sq_3}\nFourth order {r_sq_4}")
```

```python
y_pred_2 = pol2.predict(X_2_poly)
mse_2 = mean_squared_error(Y, y_pred_2)
y_pred_3 = pol3.predict(X_3_poly)
mse_3 = mean_squared_error(Y, y_pred_3)
y_pred_4 = pol4.predict(X_4_poly)
mse_4 = mean_squared_error(Y, y_pred_4)
print(f"MSE for all models\nSecond order {mse_2}\nThird order {mse_3}\nFourth order {mse_4}")
```

```python
from sklearn.pipeline import make_pipeline
```

```python
def sort_coords(X, Y):
    inds = X.values.ravel().argsort()
    x = X.values.ravel()[inds].reshape(-1,1)
    y = Y.values[inds]
    return x, y
```

```python
def fit_and_predict(order, x, y):
    poly_model = make_pipeline(PolynomialFeatures(order),
                       LinearRegression())
    poly_model.fit(x, y)
    yfit = poly_model.predict(x)
    return yfit
```

```python
x, y = sort_coords(X, Y)
yfit_2 = fit_and_predict(2, x, y)
yfit_3 = fit_and_predict(3, x, y)
yfit_4 = fit_and_predict(4, x, y)
plt.scatter(x, y,c='#ef5423', label='Scatter Plot')
plt.plot(x, yfit_2, label="Second order")
plt.plot(x, yfit_3, label="Third order")
plt.plot(x, yfit_4, label="Fourth order")
plt.legend()
```
