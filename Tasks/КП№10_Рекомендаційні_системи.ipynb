##__Рекомендаційні системи__

За останнє десятиліття Інтернет перетворився на платформу для масштабних онлайн-сервісів, що докорінно змінило підходи до комунікації та вибору. Безпрецедентна кількість товарів (ми використовуємо термін "товар" для позначення фільмів, новин, книг і продуктів), що пропонуються в Інтернеті, вимагає системи, яка може допомогти знайти товари, якому надається перевага. Таким чином, ___рекомендаційні системи___ є потужними інструментами фільтрації інформації, які можуть сприяти наданню персоналізованих послуг і надавати індивідуальний досвід для окремих користувачів.Тобто, рекомендаційні системи відіграють ключову роль у використанні багатства доступних даних, щоб зробити вибір керованим. Сьогодні підходи рекомендаційних систем лежать в основі багатьох постачальників онлайн-послуг, таких як __Amazon, Netflix__ і __YouTube__.

Рекомендаційні систем важливі для дослідників даних, оскільки вони можуть бути використані для розробки пропозицій щодо потенційно бажаних товарів.

>__Рекомендаційні системи__ - це набір алгоритмів, які використовуються для рекомендації користувачам товарів на основі інформації, отриманої від користувача.

Рекомендаційні системи використовують алгоритми машинного навчання для того, щоб робити кращі прогнози щодо вподобань користувача. Існує кілька різних алгоритмів машинного навчання, які можна використовувати в рекомендаційних системах.  Кожен алгоритм має свої сильні та слабкі сторони, і найкращий алгоритм для конкретного застосування буде залежати від характеру даних. 

##__Алгоритми машинного навчання, які можна використовувати в рекомендаційних системах__

* ___Алгоритм лінійної регресії___ використовується для пошуку найкращого лінійного наближення до набору даних. У системі рекомендацій цей алгоритм використовується для прогнозування того, як користувач оцінить товар на основі його минулих оцінок. 
* ___Нейронна мережа___  складається з взаємопов'язаних нейронів, які можуть навчитися розпізнавати шаблони. Нейронні мережі часто використовуються для задач прогнозування рекомендацій.
*___K-NearestNeighbor (KNN)___: Алгоритм KNN часто використовується для рекомендаційних систем, оскільки він здатний обробляти великі обсяги даних і може давати хороші прогнози. Алгоритм KNN працює шляхом пошуку k найближчих сусідів заданого елемента. Потім ці сусіди голосують за рейтинг товару. Алгоритм використовує середнє значення голосів для прогнозування рейтингу товару.
*___Баєсове висновування (Bayesian inference)___ - це тип алгоритму машинного навчання, який використовується для створення кращих прогнозів. Він часто використовується в рекомендаційних системах, оскільки може обробляти великі обсяги даних. Алгоритм байєсівського висновування працює за допомогою ймовірнісної моделі для прогнозування рейтингу товару. Алгоритм використовує минулі рейтинги користувача для побудови моделі ймовірності. Це дозволяє алгоритму робити кращі прогнози щодо вподобань користувача.
*___Зменшення розмірності___ - це тип алгоритму машинного навчання, який використовується для зменшення кількості вимірів у наборі даних. Він часто використовується в рекомендаційних системах, оскільки може допомогти зменшити кількість даних, які потрібно обробити. Алгоритм зменшення розмірності працює шляхом пошуку представлення даних у меншій розмірності. Це можна зробити за допомогою таких методів, як аналіз головних компонент (PCA). 


__Переваги створення рекомендаційних систем__

* Рекомендаційні системи допомагають бізнесу заробляти гроші, збільшуючи продажі та прогнозуючи, чого хочуть клієнти.
* Рекомендаційні системи можуть підвищити задоволеність клієнтів, надаючи релевантні рекомендації. Вони можуть підвищити лояльність клієнтів, пропонуючи нові товари, які можуть їм сподобатися.
* Рекомендаційні системи можуть скоротити час, необхідний для пошуку потрібного товару для покупця.
* Рекомендаційні системи можуть допомогти бізнесу дізнатися більше про вподобання своїх клієнтів.

##__Рекомендаційні системи на основі контенту (Content based, Item-Item)__

<center><img src="https://miro.medium.com/v2/resize:fit:640/format:webp/1*iTK30oDh-Vwxrk3GEmQcMA.png" width="400"></center>

Персональні рекомендації передбачають максимальне використання інформації про самого користувача, насамперед про його попередні покупки. Одним із перших з'явився підхід ___content-based filtering___. У рамках цього підходу опис товару ___(content)___ зіставляється з інтересами користувача, отриманими з його попередніх оцінок. Чим більше товар цим інтересам відповідає, тим вище оцінюється потенційна зацікавленість користувача. Очевидна вимога тут - у всіх товарів у каталозі має бути опис.

Історично предметом Content-based рекомендацій частіше були товари з неструктурованим описом: фільми, книги, статті. Такими ознаками можуть бути, наприклад, текстові описи, рецензії, склад акторів та інше. Однак ніщо не заважає використовувати і звичайні числові або категоріальні ознаки.

__Рекомендаційні системи на основі контенту__ - це тип систем рекомендацій, які надають рекомендації на основі схожості між елементами. Наприклад, якщо ви шукаєте новий фільм для перегляду, система рекомендацій на основі контенту може порекомендувати фільми, схожі на ті, які ви дивилися раніше. 

Такі рекомендаційні системи зазвичай використовуються в музиці, книгах і фільмах. Їх можна використовувати для рекомендацій продуктів, послуг або навіть веб-сайтів, вони ґрунтуються на ідеї, що якщо вам подобається один елемент, вам, швидше за все, сподобаються й інші елементи, схожі на нього. Щоб створити таку рекомендаційну систему потрібно спочатку визначити, що означає "схожість". 

Методи фільтрації на основі контенту в основному базуються на описі елемента і профілі вподобань користувача. При фільтрації на основі контенту для опису товарів використовуються ключові слова, тоді як профіль користувача створюється для того, щоб вказати тип товарів, які подобаються цьому користувачеві.

Неструктуровані ознаки описуються типовим для тексту способом - векторами в просторі слів ___(Vector-Space model)___. Кожен елемент такого вектора - ознака, що потенційно характеризує інтерес користувача. Аналогічно, продукт - вектор у тому ж просторі.

У міру взаємодії користувача із системою (скажімо, він купує фільми), векторні описи придбаних ним товарів об'єднуються (підсумовуються і нормалізуються) в єдиний вектор і, таким чином, формується вектор його інтересів. Далі достатньо знайти товар, опис якого найближчий до вектора інтересів, тобто розв'язати задачу пошуку n найближчих сусідів.

Не всі елементи однаково значущі: наприклад, союзні слова, очевидно, не несуть жодного корисного навантаження. Тому під час визначення числа елементів, що збігаються у двох векторах, усі виміри потрібно попередньо зважувати за їхньою значущістю. Це завдання вирішує добре відоме в Text Mining перетворення __TF-IDF__, яке призначає більшу вагу рідкіснішим інтересам. Збіг таких інтересів має більше значення при визначенні близькості двох векторів, ніж збіг популярних.

<center><img src="https://miro.medium.com/v2/resize:fit:1100/format:webp/1*V9ac4hLVyms79jl65Ym_Bw.jpeg" width="600"></center>

Принцип __TF-IDF__ тут тією самою мірою застосовний і до звичайних номінальних атрибутів, таких, як наприклад, жанр, режисер, мова. __TF__ - міра значущості атрибута для користувача, __IDF__ - міра "рідкості" атрибута.

> Як __міру близькості__ двох векторів найчастіше використовують косинусну відстань.

<center><img src="https://hsto.org/r/w1560/getpro/habr/post_images/0a6/136/c9b/0a6136c9bd4144fabc84b53b4e0b6f49.png" width="600"></center>


Під час додавання нової оцінки вектор інтересів оновлюється інкрементально (тільки за тими елементами, які змінилися). Під час перерахунку має сенс давати новим оцінкам трохи більшу вагу, оскільки уподобання можуть змінюватися.

```python
#Зберігаємо інформацію про фільми у фрейм даних 
movies_df = pd.read_csv('movies.csv')
#Зберігаємо інформацію про  користувача у фрейм даних
ratings_df = pd.read_csv('ratings.csv')


movies_df.head()
```

Видалимо рік зі стовпчика __title__ за допомогою функції заміни і збережемо його у новому стовпчику __year__.

```python
#Використовуємо регулярні вирази для пошуку року, що зберігається в дужках
#Вказуємо дужки, щоб не конфліктувати з фільмами, які містять рік у назві
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand=False)
#Видаляємо дужки
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand=False)
#Видалення років зі стовпця 'title'
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '')
#Застосовуємо функцію strip, щоб позбутися будь-яких пробілів, які могли з'явитися
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
movies_df.head()
```

Розділимо значення у стовпчику __"genres"__ на список жанрів, щоб спростити подальше використання. 

```python
#Кожен жанр відокремлюється символом |, потрібно викликати функцію split на |
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

Оскільки зберігання жанрів у форматі списку не є оптимальним для техніки системи рекомендацій на основі контенту, ми використаємо техніку `One Hot Encoding` для перетворення списку жанрів у вектор, де кожен стовпчик відповідає одному можливому значенню ознаки. Таке кодування необхідне для подачі категоріальних даних. У цьому випадку ми зберігаємо кожен жанр у стовпчиках, які містять або $1$, або $0$. $1$ означає, що фільм має цей жанр, а $0$ - що не має. Збережемо цей фреймворк даних в іншій змінній.

```python
#Копіювання датафрейму фільму у новий, оскільки у першому випадку нам не потрібно використовувати інформацію про жанр.

moviesWithGenres_df = movies_df.copy()

#Для кожного рядка у датафреймі перебираємо список жанрів і поміщаємо 1 у відповідний стовпчик
for index, row in movies_df.iterrows():
    for genre in row['genres']:
        moviesWithGenres_df.at[index, genre] = 1
#Заповнюємо значення NaN значенням 0, щоб показати, що фільм не має жанру у цьому стовпчику
moviesWithGenres_df = moviesWithGenres_df.fillna(0)
moviesWithGenres_df.head()
```

Розглянемо таблицю рейтингів.

```python
ratings_df.head()
```

Кожен рядок у фреймі даних про рейтинги має ідентифікатор користувача, пов'язаний принаймні з одним фільмом, рейтингом і позначкою часу, яка показує, коли він його переглянув. Нам не знадобиться стовпчик з міткою часу, тому видалимо його, щоб заощадити пам'ять.

```python
ratings_df = ratings_df.drop('timestamp', 1)
ratings_df.head()
```

> Розглянемо, як реалізувати рекомендаційну систему на основі контенту (__Item-Item__). 

Ця методика намагається з'ясувати, які аспекти елемента є улюбленими для користувача, а потім рекомендує елементи, які представляють ці аспекти. У нашому випадку ми спробуємо з'ясувати, які жанри користувач вподобав на основі фільмів та їхніх рейтингів.

Почнемо зі створення користувача, якому ми будемо рекомендувати фільми:


```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

### **Додавання `movieId` до вхідного користувача**
Після створення користувача, витягнемо ідентифікатори вхідних фільмів з фрейму даних фільмів і додамо їх до нього.

Це можна зробити спочатку відфільтрувавши рядки, що містять назви вхідних фільмів, а потім об'єднавши цю підмножину з вхідним фреймом даних. Відкинемо непотрібні стовпці для економії місця в пам'яті.


```python
#Фільтрація фільмів за назвою
inputId = movies_df[movies_df['title'].isin(inputMovies['title'].tolist())]
#Об'єднання їх, для отримання movieId. Це неявне об'єднання за назвою.
inputMovies = pd.merge(inputId, inputMovies)
#Видаляємо інформацію, яку ми не будемо використовувати, з вхідного фрейму даних
inputMovies = inputMovies.drop('genres', 1).drop('year', 1)
#Фінальний вхідний фрейм даних
#Якщо фільм, який ви додали вище, не з'явився тут, то його може не бути в оригінальному 
#dataframe або він може бути написаний по-іншому, будь ласка, перевірте написання з великої літери.
inputMovies
```

Почнемо з вивчення вподобань вхідних даних, тому візьмемо підмножину даних фільмів, які подивилися, з фрейму даних, що містить жанри, визначені двійковими значеннями.

```python
#Фільтрування фільмів з вхідних даних
userMovies = moviesWithGenres_df[moviesWithGenres_df['movieId'].isin(inputMovies['movieId'].tolist())]
userMovies
```

Знадобиться таблиця жанрів, почистимо її, скинувши індекс і видаливши стовпчики `movieId, title, genres` і `year`.

```python
#Скидання індексу
userMovies = userMovies.reset_index(drop=True)
#Видалення непотрібних значень
userGenreTable = userMovies.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
userGenreTable
```

> Починаємо вивчати вподобання вхідних даних!

Для цього перетворимо кожен жанр на вагу, використовуючи рецензії на вхідні дані, помноживши їх на таблицю жанрів, а потім підсумувавши отриману таблицю по стовпчиках. Ця операція фактично є скалярнимм добутком між матрицею та вектором.

```python
inputMovies['rating']
```

```python
#Скалярний добуток для отримання вагів вподобань у профайлі користувача
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])

userProfile
```

Використовуючи профіль користувача, ми можемо рекомендувати фільми, які задовольняють його вподобання.

```python
#Отримаємо жанри кожного фільму у нашому вихідному фреймі даних
genreTable = moviesWithGenres_df.set_index(moviesWithGenres_df['movieId'])
#Видаляємо зайву інформацію
genreTable = genreTable.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
genreTable.head()
```

Маючи профіль вхідних даних і повний список фільмів та їхніх жанрів, ми візьмемо середньозважене значення кожного фільму на основі профілю вхідних даних і порекомендуємо двадцять найкращих фільмів, які найбільше задовольняють вподобання користувача.

```python
#Множимо жанри на вагу, а потім беремо середньозважене значення
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df.head()
```
